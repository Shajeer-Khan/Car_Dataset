```python
# Dropping columns with more than 50% missing values
missing_values = data.isnull().sum()
threshold = len(data) * 0.5
data_cleaned = data.drop(columns=missing_values[missing_values > threshold].index)
```

```python
# Dropping non-informative columns
non_informative_cols = ['bulk_id', 'car_reg_no', 'latitude', 'longitude']
data_cleaned = data_cleaned.drop(columns=[col for col in non_informative_cols if col in data_cleaned])
```

```python
# Filling remaining missing values with column mean
data_cleaned.fillna(data_cleaned.mean(), inplace=True)
```

```python
# Dropping columns with no variation
constant_columns = [col for col in data_cleaned.columns if data_cleaned[col].nunique() == 1]
data_cleaned = data_cleaned.drop(columns=constant_columns)
```

```python
import matplotlib.pyplot as plt

# Generate scatter plots for each feature against the target variable
target_column = 'vehicle_speed' 
features = data_cleaned.drop(columns=[target_column])

# Plot scatter plots for each feature
for column in features.columns:
    plt.figure(figsize=(8, 6))
    plt.scatter(features[column], data_cleaned[target_column], alpha=0.5)
    plt.title(f"Scatter Plot: {column} vs {target_column}")
    plt.xlabel(column)
    plt.ylabel(target_column)
    plt.show()
```

```python
# Defining the target column and features
target_column = 'vehicle_speed' 
features = data_cleaned.drop(columns=[target_column])
target = data_cleaned[target_column]
```

```python
from sklearn.model_selection import train_test_split

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Train a Random Forest model
model = RandomForestRegressor(random_state=0)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

# Predict on the test set
y_pred = model.predict(X_test)

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
import numpy as np

# Predict on the test set
y_pred = model.predict(X_test)

# Scatter plot of actual vs predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='green', alpha=0.5, label='Random Forest Predictions')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', label='Perfect Prediction')  # Perfect prediction line
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Random Forest: Actual vs Predicted")
plt.legend()
plt.grid(True)
plt.show()
```
